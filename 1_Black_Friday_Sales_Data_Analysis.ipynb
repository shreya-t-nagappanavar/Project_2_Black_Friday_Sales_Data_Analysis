# Analyzing Gender

```python
df['Gender'].unique()
```

```python
df['Gender'] == 'M'
```

```python
c = 0                            #total number of male
for i in df['Gender']:
  if i == 'M':
    c += 1
print(c)
```

```python
c = 0                             #total number of females
for i in df['Gender']:
  if i == 'F':
    c += 1
print(c)
```

```python
df[df['Gender'] == 'M']
```

```python
len(df[df['Gender'] == 'M']),len(df[df['Gender'] == 'F'])
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']),len(df[df['Gender'] == 'F'])]}, index = ['Male' , 'Female'])
```

```python
data
```

```python
data.plot.pie(y = 'Ratio')
```

```python
data.plot.pie(y = 'Ratio' , figsize = (4,4), autopct = '%.1f')
```

```python
df.groupby('Gender').size().plot(kind = 'pie',autopct = '%.1f', title = 'Gender Ratio',figsize = (4,4))
```

```python
df.groupby('Gender').size().plot(kind = 'bar',figsize = (6,6))
```
